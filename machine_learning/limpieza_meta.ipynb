```python
import pandas as pd
import numpy as np
import openpyxl
```

```python
#importar datos

datos_meta = pd.read_excel("4_LISTA DE ALUMNOS TERMINADOS_1RA. COHORTE_MIE24MAY23.xlsx", sheet_name='ALUMNOS QUE TERMINARON', engine='openpyxl')
```

```python
display(datos_meta[0:10])
```

```python
# Unir las columnas "Nombres" y "Apellidos" con un espacio en medio
datos_meta['Nombre completo'] = datos_meta['Nombres'].astype(str) + ' ' + datos_meta['Apellidos'].astype(str)

display(datos_meta[0:10])
```

```python
# Eliminar nombres completos duplicados
datos_sin_duplicados = datos_meta.drop_duplicates(subset=['Nombre completo'])

display(datos_sin_duplicados[0:10])
```

```python
# Eliminar nombres completos duplicados
datos_sin_duplicados1 = datos_sin_duplicados.drop_duplicates(subset=['celular'])

display(datos_sin_duplicados1[0:10])
```

```python
datos_sin_duplicados1.to_excel('datos_sin_duplicados.xlsx', index=False)
```
