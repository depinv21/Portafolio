```python
# Realizar la solicitud GET al sitio web
response = requests.get(url)

# Verificar si la solicitud fue exitosa
if response.status_code == 200:
    # Obtener el contenido HTML de la respuesta
    html = response.content
    
    # Crear el objeto BeautifulSoup
    soup = BeautifulSoup(html, 'html.parser')

    # Buscar todos los elementos de tipo 'div' con la clase 'match-content ng-star-inserted'
    partidos = soup.find_all('div', class_='match-content.ng-star-inserted')
    
    # Crear una lista para almacenar los datos de los partidos
    datos_partidos = []

    # Iterar sobre los partidos y extraer los equipos y las probabilidades
    for partido in partidos:
        # Crear un diccionario para almacenar los datos del partido actual
        datos_partido = {}
        
        # Extraer los nombres de los equipos
        equipos = partido.find_all('span', class_='text truncate')
        if len(equipos) > 1:
            nombre_equipo1 = equipos[0].text.strip()
            nombre_equipo2 = equipos[1].text.strip()
            datos_partido['Equipo 1'] = nombre_equipo1
            datos_partido['Equipo 2'] = nombre_equipo2
        
        # Extraer las probabilidades (odds)
        probabilidades = partido.find_all('span', class_='bet-btn-odd')
        if len(probabilidades) > 1:
            odds_equipo1 = probabilidades[0].text.strip()
            odds_empate = probabilidades[1].text.strip()
            odds_equipo2 = probabilidades[2].text.strip()
            datos_partido['Probabilidades Equipo 1'] = odds_equipo1
            datos_partido['Probabilidades Empate'] = odds_empate
            datos_partido['Probabilidades Equipo 2'] = odds_equipo2

        # Agregar los datos del partido actual a la lista de datos de partidos
        datos_partidos.append(datos_partido)

    # Crear un DataFrame a partir de la lista de datos de partidos
    df = pd.DataFrame(datos_partidos)
    
    # Mostrar el DataFrame utilizando display
    display(df)

else:
    print("Error al realizar la solicitud:", response.status_code)
```
