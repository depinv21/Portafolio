```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
from itertools import zip_longest
import numpy as np

# URL de la página
url = 'https://espndeportes.espn.com/basquetbol/nba/calendario/_/fecha/20240423'

# Configurar el encabezado User-Agent
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'}

# Realizar la solicitud HTTP con el encabezado
response = requests.get(url, headers=headers)

# Verificar si la solicitud fue exitosa (código de estado 200)
if response.status_code == 200:
    # Parsear el contenido HTML utilizando BeautifulSoup
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Encontrar todas las celdas que contienen los equipos locales
    celdas_equipos_locales = soup.find_all('td', class_='colspan__col Table__TD')
    # Encontrar todas las celdas que contienen los equipos visitantes
    celdas_equipos_visitantes = soup.find_all('td', class_='events__col Table__TD')
    # Encontrar todas las celdas que contienen los resultados de los partidos
    celdas_resultados = soup.find_all('td', class_='teams__col Table__TD')
    
    # Extraer los nombres de los equipos locales de las celdas
    nombres_equipos_locales = [celda.find('span', class_='Table__Team').text.strip() for celda in celdas_equipos_locales]
    # Extraer los nombres de los equipos visitantes de las celdas
    nombres_equipos_visitantes = [celda.find('span', class_='Table__Team away').text.strip() for celda in celdas_equipos_visitantes]
    # Extraer los resultados de los partidos de las celdas
    resultados_partidos = [celda.find('a').text.strip() if celda.find('a') else np.nan for celda in celdas_resultados]
    
    # Combinar las listas y rellenar los valores nulos con np.nan
    data = zip_longest(nombres_equipos_locales, nombres_equipos_visitantes, resultados_partidos, fillvalue=np.nan)
    
    # Crear un DataFrame con los datos extraídos
    df = pd.DataFrame(data, columns=['Equipo Local', 'Equipo Visitante', 'Resultado'])
    
    # Mostrar el DataFrame
    print(df)
else:
    print('Error al obtener la página:', response.status_code)
```

```python
import re

# Función para procesar los resultados
def procesar_resultado(resultado):
    if resultado == 'Sin jugar':
        return pd.Series([None, None, None, None, None], index=['Equipo Ganador', 'Puntos Ganador', 'Equipo Perdedor', 'Puntos Perdedor', 'Tiempo Extra'])
    else:
        # Extraer información con expresiones regulares
        match = re.match(r'([A-Z]+) (\d+), ([A-Z]+) (\d+)( \(TE\))?', resultado)
        if match:
            equipo_ganador = match.group(1)
            puntos_ganador = int(match.group(2))
            equipo_perdedor = match.group(3)
            puntos_perdedor = int(match.group(4))
            tiempo_extra = True if match.group(5) else False
            return pd.Series([equipo_ganador, puntos_ganador, equipo_perdedor, puntos_perdedor, tiempo_extra], index=['Equipo Ganador', 'Puntos Ganador', 'Equipo Perdedor', 'Puntos Perdedor', 'Tiempo Extra'])
        else:
            return pd.Series([None, None, None, None, None], index=['Equipo Ganador', 'Puntos Ganador', 'Equipo Perdedor', 'Puntos Perdedor', 'Tiempo Extra'])

# Aplicar la función a la columna "Resultado" y concatenar los resultados a DataFrame
# Aplicar la función a la columna "Resultado" convirtiendo primero todos los valores a cadenas
df_resultados = df['Resultado'].astype(str).apply(procesar_resultado)

# Concatenar los resultados a DataFrame
df_final = pd.concat([df, df_resultados], axis=1)

# Mostrar el DataFrame final
print(df_final)
```

```python
equipo_dict = {
    'ATL': 'Atlanta',
    'BOS': 'Boston',
    'BKN': 'Brooklyn',
    'CHA': 'Charlotte',
    'CHI': 'Chicago',
    'CLE': 'Cleveland',
    'DAL': 'Dallas',
    'DEN': 'Denver',
    'DET': 'Detroit',
    'GS': 'Golden State',
    'HOU': 'Houston',
    'IND': 'Indiana',
    'LAC': 'LA Clippers',
    'LAL': 'LA Lakers',
    'MEM': 'Memphis',
    'MIA': 'Miami',
    'MIL': 'Milwaukee',
    'MIN': 'Minnesota',
    'NO': 'New Orleans',
    'NY': 'New York',
    'OKC': 'Oklahoma City',
    'ORL': 'Orlando',
    'PHI': 'Philadelphia',
    'PHX': 'Phoenix',
    'POR': 'Portland',
    'SA': 'San Antonio',
    'SAC': 'Sacramento',
    'TOR': 'Toronto',
    'UTAH': 'Utah',
    'WAS': 'Washington'
}

# Reemplazar las abreviaciones con los nombres de estado de los equipos y mantener el valor original si no está en el diccionario
df_final['Equipo Ganador'] = df_final['Equipo Ganador'].map(equipo_dict).fillna(df_final['Equipo Ganador'])
df_final['Equipo Perdedor'] = df_final['Equipo Perdedor'].map(equipo_dict).fillna(df_final['Equipo Perdedor'])

# Mostrar el DataFrame actualizado
print(df_final)
```

```python
# Realizar las sustituciones en las columnas "Equipo Local" y "Equipo Visitante"
df_final['Equipo Local'] = df_final['Equipo Local'].replace({'LA': 'LA Clippers', 'Los Angeles': 'LA Lakers'})
df_final['Equipo Visitante'] = df_final['Equipo Visitante'].replace({'LA': 'LA Clippers', 'Los Angeles': 'LA Lakers'})
```

```python
# Crear nuevas columnas "C" y "V" y asignar los valores correspondientes
df_final['C'] = df_final.apply(lambda row: row['Puntos Ganador'] if row['Equipo Local'] == row['Equipo Ganador'] else row['Puntos Perdedor'], axis=1)
df_final['V'] = df_final.apply(lambda row: row['Puntos Perdedor'] if row['Equipo Local'] == row['Equipo Ganador'] else row['Puntos Ganador'], axis=1)

# Mostrar el DataFrame actualizado
print(df_final)
```
