```python
# Importar las bibliotecas necesarias
import pandas as pd
from bs4 import BeautifulSoup
import requests
from datetime import datetime, timedelta

# Función para generar las fechas de inicio y finalización de la temporada
def generar_fechas_temporada():
    fecha_inicio = datetime(2023, 10, 5)  # Fecha de inicio de la temporada
    fecha_finalizacion = datetime(2024, 5, 5)  # Fecha de finalización de la temporada
    return fecha_inicio, fecha_finalizacion

# Función para generar las URL de las jornadas de la temporada
def generar_urls_jornadas_temporada():
    fecha_inicio, fecha_finalizacion = generar_fechas_temporada()
    urls_jornadas = []
    fecha_actual = fecha_inicio
    while fecha_actual <= fecha_finalizacion:
        url_jornada = f'https://espndeportes.espn.com/basquetbol/nba/calendario/_/fecha/{fecha_actual.strftime("%Y%m%d")}'
        urls_jornadas.append(url_jornada)
        fecha_actual += timedelta(days=1)
    return urls_jornadas

# Configurar el encabezado User-Agent
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'}

# Lista para almacenar los DataFrames de cada jornada
dfs_jornadas = []

# Iterar sobre las URLs de las jornadas de la temporada
for url_jornada in generar_urls_jornadas_temporada():
    print(f'Extrayendo datos de la jornada: {url_jornada}')
    # Realizar la solicitud HTTP con el encabezado
    response = requests.get(url_jornada, headers=headers)

    # Verificar si la solicitud fue exitosa (código de estado 200)
    if response.status_code == 200:
        # Parsear el contenido HTML utilizando BeautifulSoup
        soup = BeautifulSoup(response.content, 'html.parser')

        # Encontrar todas las celdas que contienen los equipos locales
        celdas_equipos_locales = soup.find_all('td', class_='colspan__col Table__TD')
        # Encontrar todas las celdas que contienen los equipos visitantes
        celdas_equipos_visitantes = soup.find_all('td', class_='events__col Table__TD')
        # Encontrar todas las celdas que contienen los resultados de los partidos
        celdas_resultados = soup.find_all('td', class_='teams__col Table__TD')

        # Extraer los nombres de los equipos locales de las celdas
        nombres_equipos_locales = [celda.find('span', class_='Table__Team').text.strip() for celda in celdas_equipos_locales]
        # Extraer los nombres de los equipos visitantes de las celdas
        nombres_equipos_visitantes = [celda.find('span', class_='Table__Team away').text.strip() for celda in celdas_equipos_visitantes]
        # Extraer los resultados de los partidos de las celdas
        resultados_partidos = [celda.find('a').text.strip() if celda.find('a') else None for celda in celdas_resultados]

        # Reemplazar valores vacíos en la columna de Resultado con "Sin resultado"
        resultados_partidos = ['Sin resultado' if resultado is None else resultado for resultado in resultados_partidos]

        # Crear un DataFrame con los datos extraídos
        if len(nombres_equipos_locales) == len(nombres_equipos_visitantes):
            if len(resultados_partidos) < len(nombres_equipos_locales):
                resultados_partidos.extend(['Sin jugar'] * (len(nombres_equipos_locales) - len(resultados_partidos)))

            df_jornada = pd.DataFrame({
                'Equipo Local': nombres_equipos_locales,
                'Equipo Visitante': nombres_equipos_visitantes,
                'Resultado': resultados_partidos
            })

            # Agregar el DataFrame de la jornada a la lista
            dfs_jornadas.append(df_jornada)
            print(f'Datos de la jornada extraídos correctamente. Longitud de los datos: Equipos Locales: {len(nombres_equipos_locales)}, Equipos Visitantes: {len(nombres_equipos_visitantes)}, Resultados: {len(resultados_partidos)}')
        else:
            print(f'Error: Las listas de equipos locales y visitantes tienen diferentes longitudes. Equipos Locales: {len(nombres_equipos_locales)}, Equipos Visitantes: {len(nombres_equipos_visitantes)}. Se omitirá esta jornada.')

# Concatenar todos los DataFrames de las jornadas en uno solo
df_temporada = pd.concat(dfs_jornadas, ignore_index=True)

# Mostrar el DataFrame de la temporada
print(df_temporada)
```

```python
import re

# Función para procesar los resultados
def procesar_resultado(resultado):
    if resultado == 'Sin jugar':
        return pd.Series([None, None, None, None, None], index=['Equipo Ganador', 'Puntos Ganador', 'Equipo Perdedor', 'Puntos Perdedor', 'Tiempo Extra'])
    else:
        # Extraer información con expresiones regulares
        match = re.match(r'([A-Z]+) (\d+), ([A-Z]+) (\d+)( \(TE\))?', resultado)
        if match:
            equipo_ganador = match.group(1)
            puntos_ganador = int(match.group(2))
            equipo_perdedor = match.group(3)
            puntos_perdedor = int(match.group(4))
            tiempo_extra = True if match.group(5) else False
            return pd.Series([equipo_ganador, puntos_ganador, equipo_perdedor, puntos_perdedor, tiempo_extra], index=['Equipo Ganador', 'Puntos Ganador', 'Equipo Perdedor', 'Puntos Perdedor', 'Tiempo Extra'])
        else:
            return pd.Series([None, None, None, None, None], index=['Equipo Ganador', 'Puntos Ganador', 'Equipo Perdedor', 'Puntos Perdedor', 'Tiempo Extra'])

# Aplicar la función a la columna "Resultado" y concatenar los resultados a DataFrame
df_resultados = df_temporada['Resultado'].apply(procesar_resultado)
df_final = pd.concat([df_temporada, df_resultados], axis=1)

# Mostrar el DataFrame final
print(df_final)
```

```python
equipo_dict = {
    'ATL': 'Atlanta',
    'BOS': 'Boston',
    'BKN': 'Brooklyn',
    'CHA': 'Charlotte',
    'CHI': 'Chicago',
    'CLE': 'Cleveland',
    'DAL': 'Dallas',
    'DEN': 'Denver',
    'DET': 'Detroit',
    'GS': 'Golden State',
    'HOU': 'Houston',
    'IND': 'Indiana',
    'LAC': 'LA Clippers',
    'LAL': 'LA Lakers',
    'MEM': 'Memphis',
    'MIA': 'Miami',
    'MIL': 'Milwaukee',
    'MIN': 'Minnesota',
    'NO': 'New Orleans',
    'NY': 'New York',
    'OKC': 'Oklahoma City',
    'ORL': 'Orlando',
    'PHI': 'Philadelphia',
    'PHX': 'Phoenix',
    'POR': 'Portland',
    'SA': 'San Antonio',
    'SAC': 'Sacramento',
    'TOR': 'Toronto',
    'UTAH': 'Utah',
    'WAS': 'Washington'
}

# Reemplazar las abreviaciones con los nombres de estado de los equipos y mantener el valor original si no está en el diccionario
df_final['Equipo Ganador'] = df_final['Equipo Ganador'].map(equipo_dict).fillna(df_final['Equipo Ganador'])
df_final['Equipo Perdedor'] = df_final['Equipo Perdedor'].map(equipo_dict).fillna(df_final['Equipo Perdedor'])

# Mostrar el DataFrame actualizado
print(df_final)
```

```python
# Realizar las sustituciones en las columnas "Equipo Local" y "Equipo Visitante"
df_final['Equipo Local'] = df_final['Equipo Local'].replace({'LA': 'LA Clippers', 'Los Angeles': 'LA Lakers'})
df_final['Equipo Visitante'] = df_final['Equipo Visitante'].replace({'LA': 'LA Clippers', 'Los Angeles': 'LA Lakers'})
```

```python
# Crear nuevas columnas "C" y "V" y asignar los valores correspondientes
df_final['C'] = df_final.apply(lambda row: row['Puntos Ganador'] if row['Equipo Local'] == row['Equipo Ganador'] else row['Puntos Perdedor'], axis=1)
df_final['V'] = df_final.apply(lambda row: row['Puntos Perdedor'] if row['Equipo Local'] == row['Equipo Ganador'] else row['Puntos Ganador'], axis=1)

# Mostrar el DataFrame actualizado
print(df_final)
```
