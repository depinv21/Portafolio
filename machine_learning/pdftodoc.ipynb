```python
import PyPDF2
from docx import Document
from translate import Translator
```

```python
def convert_pdf_to_docx(input_pdf, output_docx, target_language, start_page=None, end_page=None):
    pdf_reader = PyPDF2.PdfReader(input_pdf)
    document = Document()

    if start_page is None:
        start_page = 0
    else:
        start_page = max(0, start_page - 1)

    if end_page is None:
        end_page = len(pdf_reader.pages)
    else:
        end_page = min(end_page, len(pdf_reader.pages))

    translator = Translator(to_lang=target_language)

    for page_num in range(start_page, end_page):
        page = pdf_reader.pages[page_num]
        text = page.extract_text()

        translation = translator.translate(text)

        paragraph = document.add_paragraph()
        paragraph.add_run(text)
        paragraph.add_run('\n')
        paragraph.add_run(translation)
        paragraph.add_run('\n')

    document.save(output_docx)
```

```python
# Ejemplo de uso
input_pdf = r'C:\Users\David Pineda\Desktop\libro machine learning\Hands-On-Machine-Learning-with-Scikit-Learn-and-TensorFlow.pdf'
  # Ruta al archivo PDF de entrada
output_docx = r'C:\Users\David Pineda\Desktop\libro machine learning\librodoc.docx'  # Ruta al archivo DOCX de salida
target_language = 'es'  # Idioma objetivo para la traducción
start_page = 29  # Página de inicio para la traducción (opcional)
end_page = 61  # Página de fin para la traducción (opcional)
```

```python
convert_pdf_to_docx(input_pdf, output_docx, target_language, start_page, end_page)
```
