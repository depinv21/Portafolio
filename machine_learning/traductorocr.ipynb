```python
import PyPDF2
from pdf2image import convert_from_path
import pytesseract
import pytesseract
from PIL import Image
from googletrans import Translator
```

```python
def extract_text_from_image(image):
    text = pytesseract.image_to_string(image)
    return text

def translate_pdf(input_file, output_file, target_language, start_page=None, end_page=None):
    translator = Translator()

    images = convert_from_path(input_file)

    if start_page is None:
        start_page = 0
    else:
        start_page = max(0, start_page - 1)  # Ajustar el número de página base 1 a base 0

    if end_page is None:
        end_page = len(images)
    else:
        end_page = min(end_page, len(images))

    pdf_writer = PyPDF2.PdfWriter()

    for page_num in range(start_page, end_page):
        image = images[page_num]

        text = extract_text_from_image(image)

        if text:
            translation = translator.translate(text, dest=target_language)

            if translation.text:
                translated_page = PyPDF2.PageObject.create_blank_page(None, image.width, image.height)
                translated_page.merge_page(image)
                translated_page.title = translation.text
                pdf_writer.add_page(translated_page)
                print(f"Traduciendo página {page_num + 1}/{len(images)}")
            else:
                print(f"No se pudo traducir la página {page_num + 1}")
        else:
            print(f"No se encontró texto en la página {page_num + 1}")

    with open(output_file, 'wb') as output:
        pdf_writer.write(output)
```

```python
# Ruta del archivo PDF de entrada
input_pdf = r'C:\Users\David Pineda\Desktop\libro machine learning\Hands-On-Machine-Learning-with-Scikit-Learn-and-TensorFlow.pdf'

# Ruta del archivo PDF de salida (traducido)
output_pdf = r'C:\Users\David Pineda\Desktop\libro machine learning\traducido.pdf'

# Idioma objetivo de la traducción (por ejemplo, 'es' para español)
target_language = 'es'
```

```python
# Llamar a la función para traducir el PDF
translate_pdf(input_pdf, output_pdf, 'es', start_page=29, end_page=61)
```
