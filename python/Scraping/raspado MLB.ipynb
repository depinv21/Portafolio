```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
import numpy as np
from datetime import datetime, timedelta

# Función para formatear la fecha en el formato de la URL
def format_date(date):
    return date.strftime("%Y%m%d")

# Lista para almacenar los DataFrames de cada día
dfs = []

# Rango de fechas
start_date = datetime(2024, 3, 20)
end_date = datetime(2024, 9, 19)
delta = timedelta(days=1)  # Incremento diario

# Configurar el encabezado User-Agent
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'}

# Iterar sobre el rango de fechas
current_date = start_date
while current_date <= end_date:
    # Construir la URL para la fecha actual
    url = f'https://espndeportes.espn.com/beisbol/mlb/calendario/_/fecha/{format_date(current_date)}'

    # Realizar la solicitud HTTP con el encabezado
    response = requests.get(url, headers=headers)

    # Verificar si la solicitud fue exitosa (código de estado 200)
    if response.status_code == 200:
        # Parsear el contenido HTML utilizando BeautifulSoup
        soup = BeautifulSoup(response.content, 'html.parser')

        # Encontrar todas las celdas que contienen el equipo local
        celdas_equipos_locales = soup.find_all('td', class_='colspan__col Table__TD')
        nombres_equipos_locales = [celda.find('span', class_='Table__Team').text.strip() for celda in celdas_equipos_locales]

        # Encontrar todas las celdas que contienen el equipo visitante
        celdas_equipos_visitantes = soup.find_all('td', class_='events__col Table__TD')
        nombres_equipos_visitantes = [celda.find('span', class_='Table__Team away').text.strip() for celda in celdas_equipos_visitantes]

        # Encontrar todas las celdas que contienen los resultados de los partidos
        celdas_resultados = soup.find_all('td', class_='teams__col Table__TD')
        resultados_partidos = []
        for celda in celdas_resultados:
            resultado_partido_tag = celda.find('a')
            if resultado_partido_tag:
                resultado_partido = resultado_partido_tag.text.strip()
                if any(char.isdigit() for char in resultado_partido) or "Cancelado" in resultado_partido:
                 resultados_partidos.append(resultado_partido)

        # Determinar la longitud máxima de las listas
        max_length = max(len(nombres_equipos_locales), len(nombres_equipos_visitantes), len(resultados_partidos))

        # Rellenar las listas con NaN hasta alcanzar la longitud máxima
        nombres_equipos_locales += [np.nan] * (max_length - len(nombres_equipos_locales))
        nombres_equipos_visitantes += [np.nan] * (max_length - len(nombres_equipos_visitantes))
        resultados_partidos += [np.nan] * (max_length - len(resultados_partidos))

        # Crear el DataFrame para la fecha actual
        df = pd.DataFrame({
            'Equipo Local': nombres_equipos_locales,
            'Equipo Visitante': nombres_equipos_visitantes,
            'Resultado': resultados_partidos
        })

        # Agregar el DataFrame al listado
        dfs.append(df)
    else:
        print('Error al obtener la página:', response.status_code)

    # Avanzar al siguiente día
    current_date += delta

# Concatenar todos los DataFrames en uno solo
result_df = pd.concat(dfs, ignore_index=True)

# Imprimir el DataFrame completo
print(result_df)
```

```python
# Función para procesar los resultados
def procesar_resultado(resultado):
    if resultado == 'Sin jugar':
        return pd.Series([None, None, None, None], index=['Equipo Ganador', 'Puntos Ganador', 'Equipo Perdedor', 'Puntos Perdedor'])
    else:
        # Extraer información con expresiones regulares
        match = re.match(r'([A-Z]+) (\d+), ([A-Z]+) (\d+)( \(TE\))?', resultado)
        if match:
            equipo_ganador = match.group(1)
            puntos_ganador = int(match.group(2))
            equipo_perdedor = match.group(3)
            puntos_perdedor = int(match.group(4))
            return pd.Series([equipo_ganador, puntos_ganador, equipo_perdedor, puntos_perdedor], index=['Equipo Ganador', 'Puntos Ganador', 'Equipo Perdedor', 'Puntos Perdedor'])
        else:
            return pd.Series([None, None, None, None], index=['Equipo Ganador', 'Puntos Ganador', 'Equipo Perdedor', 'Puntos Perdedor'])

# Aplicar la función a la columna "Resultado" y concatenar los resultados a DataFrame
df_resultados = df_temporada['Resultado'].apply(procesar_resultado)
df_final = pd.concat([df_temporada, df_resultados], axis=1)

# Mostrar el DataFrame final
print(df_final)
```

```python
# Equipos ganadores de las ciudades especificadas
equipos_ganadores_ciudades = {
    'Los Angeles': ['Los Angeles Angels', 'Los Angeles Dodgers'],
    'Chicago': ['Chicago Cubs', 'Chicago White Sox'],
    'New York': ['New York Mets', 'New York Yankees']
}

# Función para determinar el equipo local basado en las condiciones especificadas
def determinar_equipo_local(row):
    equipo_local = row['Equipo Local']
    equipo_ganador = row['Equipo Ganador']
    equipo_perdedor = row['Equipo Visitante'] if equipo_local == row['Equipo Local'] else equipo_local
    
    # Verificar si el equipo local es Los Angeles y el equipo ganador es uno de los equipos de Los Angeles
    if equipo_local == 'Los Angeles' and equipo_ganador in equipos_ganadores_ciudades['Los Angeles']:
        return equipo_ganador
    
    # Verificar si el equipo local es Chicago y el equipo ganador es uno de los equipos de Chicago
    if equipo_local == 'Chicago' and equipo_ganador in equipos_ganadores_ciudades['Chicago']:
        return equipo_ganador
    
    # Verificar si el equipo local es New York y el equipo ganador es uno de los equipos de New York
    if equipo_local == 'New York' and equipo_ganador in equipos_ganadores_ciudades['New York']:
        return equipo_ganador
    
    # Si el equipo local es Los Angeles pero el equipo ganador no es uno de los equipos de Los Angeles, asignar el equipo perdedor como el nuevo equipo local
    if equipo_local == 'Los Angeles' and equipo_perdedor not in equipos_ganadores_ciudades['Los Angeles']:
        return equipo_perdedor
    
    # Si el equipo local es Chicago pero el equipo ganador no es uno de los equipos de Chicago, asignar el equipo perdedor como el nuevo equipo local
    if equipo_local == 'Chicago' and equipo_perdedor not in equipos_ganadores_ciudades['Chicago']:
        return equipo_perdedor
    
    # Si el equipo local es New York pero el equipo ganador no es uno de los equipos de New York, asignar el equipo perdedor como el nuevo equipo local
    if equipo_local == 'New York' and equipo_perdedor not in equipos_ganadores_ciudades['New York']:
        return equipo_perdedor
    
    return equipo_local

# Aplicar la función a cada fila del DataFrame
df_final['Equipo Local'] = df_final.apply(determinar_equipo_local, axis=1)

# Imprimir el DataFrame final
print(df_final)
```
